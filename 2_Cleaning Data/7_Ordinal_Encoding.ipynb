```python
import pandas as pd
```

```python
df = pd.DataFrame({"size":["s","m","l","xl","s", "m","xl","l","m","s","xl","m"]})
df
```

```python
 df.head(4)
```

# Using SCKIT Learn

```python
# Understand or analyse category and provide in single variable
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)  # oe is object
oe.fit(df[["size"]])
```

```python
df["size_enco"] = oe.transform(df[["size"]])
```

```python
df
```

# Using MAP FUNCTION

```python
ord_data1 = {"s":0, "m":1, "l":2, "xl":3}
```

```python
df["size_using_map"]= df["size"].map(ord_data1)
```
