```python
# Forwardfilling with column
df.fillna(method="ffill", axis=1) # cheak last entry of Ca and try to change axis
```

```python
# Backwardfilling with column
df.fillna(method="bfill", axis=1)
```

```python
# Forwardfilling with row
df.fillna(method="ffill", axis=0)
```

```python
# Backwardfilling with column
df.fillna(method="bfill", axis=1)
```

```python
# Mod filling means null data filled with most frequently accured data
# 1. find mod
df["Ca"].mode()
```

```python
# 2. above is Series dataset so we want it in direct value thus:
df["Ca"].mode()[0]
```

```python
# 3. Apply it to Ca column
df["Ca"].fillna(df["Ca"].mode()[0], inplace = True)
```

```python
# To select all cloumns havinh float as datatype
df.select_dtypes(include="float").columns
```

```python
# To provide mod value to all columns having float as data type
# Only missing content will filled
for i in df.select_dtypes(include="float").columns:
    df[i].fillna(df[i].mode()[0], inplace = True)
    
```
