```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("collegePlace.csv")
```

```python
df.head()
```

```python
x = df[["CGPA"]]
y = df["PlacedOrNot"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Voting Classifier

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
dt.score(x_train, y_train)*100 , dt.score(x_test, y_test)*100
```

```python
dt2 = LinearRegression()
dt2.fit(x_train, y_train)
dt2.score(x_train, y_train)*100 , dt2.score(x_test, y_test)*100
```

```python
dt1 = SVR()
dt1.fit(x_train, y_train)
dt1.score(x_train, y_train)*100 , dt1.score(x_test, y_test)*100
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
li = [("dt", DecisionTreeRegressor()),("dt2",LinearRegression()),("dt1",SVR())]
```

```python
vc = VotingRegressor(li)
vc.fit(x_train, y_train)
```

```python
vc.score(x_train,y_train)*100
```

```python
vc.score(x_test, y_test)*100
```

```python
pred = {"dt":dt.predict(x_test), "dt1":dt1.predict(x_test), "dt2":dt2.predict(x_test),
        "vc":vc.predict(x_test)}
```
