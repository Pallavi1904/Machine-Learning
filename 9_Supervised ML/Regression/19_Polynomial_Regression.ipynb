```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df.head()
```

```python
plt.scatter(df["BloodPressure"],df["Age"])
plt.xlabel("BloodPressure")
plt.ylabel("Age")
plt.show()
```

```python
df.corr()
#Cheak of blood pressure and age
```

```python
x = df[["BloodPressure"]]
x
```

```python
y = df["Age"]
y
```

```python
from sklearn.preprocessing import PolynomialFeatures #To fit data in polynomial
```

```python
pf = PolynomialFeatures(degree=2)
```

```python
pf.fit(x)
x = pf.transform(x) #Again put in x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=34)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
pred = lr.predict(x)
```

```python
plt.scatter(df["BloodPressure"],df["Age"])
plt.plot(df["BloodPressure"], pred,c='red')
plt.xlabel("BloodPressure")
plt.ylabel("Age")
plt.show()
```
