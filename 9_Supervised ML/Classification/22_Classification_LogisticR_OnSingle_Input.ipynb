```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"diabetes.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(columns=['Pregnancies', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age'],inplace=True)
```

```python
df
```

```python
sns.scatterplot(x="Glucose", y="Outcome", data=df)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df[["Glucose"]]
```

```python
y = df["Outcome"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[148]]) #If glucose is 148 ?
```

```python
sns.scatterplot(x="Glucose", y="Outcome", data=df)
sns.lineplot(x ="Glucose", y=lr.predict(x),data=df, color='red')
plt.show()
```
