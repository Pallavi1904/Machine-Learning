```python
market = []
for i in range(0, df.shape[0]):
    cus = []
    for j in df.columns:
        if type(df[j][i])==str:
            cus.append(df[j][i])
    market.append(cus)
    
```

```python
l = []
for i in market:
    for j in i:
        l.append(j)
```

# Transaction Matrix Formation

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
t = TransactionEncoder()
```

```python
t.fit(market)
```

```python
new_df = pd.DataFrame(t.transform(market), columns=t.columns_)
```

```python
new_df
```

# Aprior Algorithm

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(new_df, min_support=0.07,use_colnames=True, max_len=3)
```

```python
apriori(new_df, min_support=0.05,use_colnames=True, max_len=3).sort_values(by=["support"])
```
