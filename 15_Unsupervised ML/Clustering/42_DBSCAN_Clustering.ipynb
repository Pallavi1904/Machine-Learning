```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=250,noise=0.05)
```

```python
x
```

```python
 y
```

```python
df = {"data1":x[:,0], "data2":x[:,1]}
```

```python
dataset = pd.DataFrame(df)
```

```python
dataset.head()
```

```python
sns.scatterplot(x="data1", y="data2",data=dataset)
plt.show()
# Non linear data
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.2,min_samples=5) # by taking 5 (chaek once output by radius 0.5)
```

```python
dataset["predict"] = db.fit_predict(dataset)
```

```python
sns.scatterplot(x="data1", y="data2",data= dataset, hue="predict")
plt.show()
```
