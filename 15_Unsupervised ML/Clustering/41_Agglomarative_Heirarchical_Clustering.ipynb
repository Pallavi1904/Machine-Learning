```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.head(2)
```

```python
sns.pairplot(data= df)
plt.show()
# Observe data whether it is linealy separable or not
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,20))
sc.dendrogram(sc.linkage(df,method='single', metric='euclidean'))
plt.show()
# IT making 2 clusters
```

```python
plt.figure(figsize=(20,20))
sc.dendrogram(sc.linkage(df,method='complete', metric='euclidean'))
plt.show()
# It also making 2 clusters
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
sg = AgglomerativeClustering(n_clusters=2,linkage='single')
```

```python
df["predict"] = sg.fit_predict(df)
```

```python
df.head()
```

```python
sns.pairplot(data=df, hue="predict")
plt.show()
```
