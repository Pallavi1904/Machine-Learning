```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv(r"Social_Network_Ads.csv")
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
```

```python
dt.fit(x_train, y_train)
```

```python
dt.score(x_test, y_test)*100
```

```python
dt.score(x_train, y_train)*100
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
dt = {"criterion" : ["squared_error", "friedman_mse", "absolute_error","poisson"],
      "splitter" : ["best","random"],
    "max_depth" : [i for i in range(2,20)]}
```

# GridSearchCV

```python
gd = GridSearchCV(DecisionTreeRegressor(), param_grid=dt)
```

```python
gd.fit(x_train, y_train)
```

```python
gd.best_params_ #provides best hyperparameter that give highest accuracy together
```

```python
gd.best_score_
```

```python
dt1 = DecisionTreeRegressor(criterion="poisson",max_depth=4,splitter="best")
```

```python
dt1.fit(x_train, y_train)
```

```python
dt1.score(x_test, y_test)*100
```

```python
dt1.score(x_train, y_train)*100
```

# RandomisedSearchCV

```python
rd = RandomizedSearchCV(DecisionTreeRegressor(), param_distributions=dt, n_iter=20)
```

```python
rd.fit(x_train, y_train)
```
