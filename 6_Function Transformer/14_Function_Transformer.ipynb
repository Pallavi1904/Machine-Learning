```python
 import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("temperatures.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.distplot(df["ANNUAL"])
plt.show()
```

```python
q3 = df["ANNUAL"].quantile(0.75)
q1 = df["ANNUAL"].quantile(0.25)
```

```python
iqr = q3 - q1
iqr
```

```python
max_range = q3 + (1.5*iqr)
max_range
```

```python
df = df["ANNUAL"] <= max_range
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func=np.log1p)
```

```python
df = df.reset_index()  # Move index to columns
ft.fit(df[["ANNUAL"]])
```

```python
df["ANNUAL_ft"] = ft.transform(df[["ANNUAL"]])
```

```python
sns.distplot(df["ANNUAL"])
plt.show()
```
