```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv(r"Social_Network_Ads.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(9,7))
sns.scatterplot(x="Age", y ="EstimatedSalary",data = df,hue="Purchased")
plt.show()
```

```python
x = df[["Age","EstimatedSalary"]]
y = df["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.svm import SVC #SVC = Support Vector Classifier
```

# Linear Kernel Function

```python
sv = SVC(kernel="linear")
```

```python
sv.fit(x_train, y_train)
```

```python
sv.score(x_test, y_test)*100
```

```python
sv.score(x_train, y_train)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=sv)
plt.show()
```

# Polynomial Kernel Function

```python
sv1 = SVC(kernel="poly")
sv1.fit(x_train, y_train)
```

```python
sv.score(x_test, y_test)*100
```

```python
sv.score(x_train, y_train)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=sv1)
plt.show()
```

# Sigmoid Kernel Function

```python
sv2 = SVC(kernel="sigmoid")
sv2.fit(x_train, y_train)
```

```python
sv.score(x_test, y_test)*100
```

```python
sv.score(x_train, y_train)*100
```
